```python
data=data[data.Cuisines.isna()==False]
```

```python
data.dtypes
```

```python
sns.countplot(x='Price_range',data=data,color='orange')
plt.show()
```

```python
sns.scatterplot(x='Average_Cost_for_two',y='Rating',data=data)
plt.show()
```

```python
sns.boxplot(
x='Has_Online_delivery',
y='Rating',
data=data
)
plt.show()
```

```python
corr=data.corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(
corr,
annot=True,
cmap='coolwarm'
)

plt.show()
```

```python
Top_city=data['City'].value_counts().head()
Top_city.plot(kind='bar',color='red')
```

```python
Top_rating_by_city=data.groupby('Rating')['City'].value_counts().head()
Top_rating_by_city.plot(kind='bar')
```

```python
Top_Cuisines=data['Cuisines'].value_counts().head(10)
Top_Cuisines.plot(kind='line')
```

```python
top_restaurants=data.sort_values(
by='Rating',
ascending=False
)

print(top_restaurants[
['RestaurantName',
'Rating']
].head(10))
```

```python
Delivery = data['Has_Online_delivery'].value_counts().head()

plt.pie(
    Delivery,
    labels=Delivery.index,
    autopct='%1.1f%%'
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
x=data['Average_Cost_for_two']
)

plt.show()
```

```python
data['Cost_Category']=pd.cut(
data['Average_Cost_for_two'],
bins=[0,500,1000,3000,100000],
labels=['Low','Medium','High','Luxury']
)
```

```python
print(data[['Average_Cost_for_two','Cost_Category']].head(10))
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Cost_Category',
    data=data
)

plt.show()
```
